```python
y=df.LogOfPrice
y
```

```python
dftemp=df.select_dtypes(include=['int64','float64'])
dftemp
```

```python
x=dftemp.drop(['LogOfPrice'],axis=1)
x
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.5,random_state=3)
```

```python
X_train
```

```python
Y_train
```

```python
linear=LinearRegression()
linear.fit(X_train, Y_train)
```

```python
result=linear.predict(X_test)
```

```python
result_score=linear.score(X_test, Y_test)
print("Accuracy:",result_score)
```

```python
linear_cross=cross_val_score(linear,x,y,cv=5,scoring='r2')
```

```python
print("Cross Validation Result:",linear_cross)
print("R2",linear_cross.mean())
```
